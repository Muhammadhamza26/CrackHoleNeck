```python
import os
import random
import shutil

# Set the paths
train_img_path = "./crack_data/Dataset/images/train"
val_img_path = "./crack_data/Dataset/images/val"

train_label_path = "./crack_data/Dataset/labels/train"
val_label_path = "./crack_data/Dataset/labels/val"

# Get the list of all image files in the source directory
image_files = [
    file
    for file in sorted(os.listdir(train_img_path))
    if file.endswith(".jpg") or file.endswith(".png")
]

# Calculate the number of images to move
percent = 0.20
num_images_to_move = int(len(image_files) * percent)

# Randomly select the images to move
images_to_move = random.sample(image_files, num_images_to_move)

print(f"{num_images_to_move} images to move.")

# Move the selected images to the validation directory
for image in images_to_move:
    image_base = image.split(".")[0]
    # move images
    shutil.move(os.path.join(train_img_path, image), os.path.join(val_img_path, image))
    print(f"Image {image} moved.")
    # move labels
    text_file = f"{image_base}.txt"
    shutil.move(
        os.path.join(train_label_path, text_file), os.path.join(val_label_path, text_file)
    )
    print(f"## Label {text_file} moved.")

```

```python
import numpy

def resize(image, input_size):
    shape = image.shape[:2]  # current shape [height, width]
    r = min(1.0, input_size / shape[0], input_size / shape[1])
    pad = int(round(shape[1] * r)), int(round(shape[0] * r))

    w = numpy.mod(input_size - pad[0], 32) / 2
    h = numpy.mod(input_size - pad[1], 32) / 2

    if shape[::-1] != pad:  # resize
        image = cv2.resize(image, dsize=pad, interpolation=cv2.INTER_LINEAR)
    top, bottom = int(round(h - 0.1)), int(round(h + 0.1))
    left, right = int(round(w - 0.1)), int(round(w + 0.1))
    image = cv2.copyMakeBorder(image, top, bottom, left, right, cv2.BORDER_CONSTANT)

    # Convert BGR to GRAY
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    image = numpy.stack((image, image, image), axis=-1)
    return image

# def get_data(input_size=640):
#     ROOT = "/home/sajid/cctv/YOLOv8-crackV7/samples_in_jpg/"
#     OUT_DIR = "/home/sajid/cctv/YOLOv8-crackV7/offline_infer/"
#     dataset_loader = []
#     for file in sorted(os.listdir(ROOT)):
#         if file.endswith(".jpg"):
#             file_path = os.path.join(ROOT, file)
#             image = cv2.imread(file_path)
#             x = np.stack([resize(i, input_size) for i in image], axis=0)
#             x = x.transpose((0, 3, 1, 2))  # HWC to CHW
#             x = np.ascontiguousarray(x)  # contiguous
#             dataset_loader.append(file_path)
#     return dataset_loader
# data = get_data()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import numpy as np
import random
import os

# Function to parse bounding box information from the file
def parse_bboxes(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    bboxes = [list(map(float, line.strip().split(' '))) for line in lines]
    return bboxes

# Function to convert normalized coordinates to pixel coordinates
def convert_to_pixel_coords(bbox, img_width, img_height):
    class_id, x_center, y_center, width, height = bbox
    x_center, width = x_center * img_width, width * img_width
    y_center, height = y_center * img_height, height * img_height
    x_min = int(x_center - width / 2)
    y_min = int(y_center - height / 2)
    return int(class_id), x_min, y_min, int(width), int(height)

# Function to draw bounding boxes and labels using matplotlib
def draw_bboxes(image_path, bboxes):
    # Load image
    img = Image.open(image_path).convert("RGB")
    # img_rgb = img_bgr
    img_width, img_height = img.size

    # Create figure and axes
    fig, ax = plt.subplots(1)

    # Display the image
    ax.imshow(img)

    # Convert and draw each bounding box
    for bbox in bboxes:
        class_id, x_min, y_min, width, height = convert_to_pixel_coords(bbox, img_width, img_height)
        if class_id == 0:
            rect = patches.Rectangle((x_min, y_min), width, height, linewidth=2, edgecolor='g', facecolor='none')
        else:
            rect = patches.Rectangle((x_min, y_min), width, height, linewidth=2, edgecolor='m', facecolor='none')
        ax.add_patch(rect)
        # plt.text(x_min, y_min, str(class_id), color='white', fontsize=12, backgroundcolor='red')

    plt.show()

def get_list_of_data(path, fileformat):
    """_summary_

    Args:
        path (_str_): Path of the data directory.
        fileformat (_str_): Format of the data to be retreived, e.g., "txt", "jpg", "tiff", etc. 

    Returns:
        _list_: A list contains all images path.
    """
    data_path = [os.path.join(path, f) for f in sorted(os.listdir(path)) if f.endswith(fileformat)]
    return data_path

# # Example usage
labels_path = "/hdd/KIA_Smart_Factory_Crack_Hole_detection/sajid/v5/labels/train/"
images_path = "/hdd/KIA_Smart_Factory_Crack_Hole_detection/sajid/v5/images/train/"
# List of images/labels paths
label_files = get_list_of_data(labels_path, "txt")
image_files = get_list_of_data(images_path, "png")


# Draw bounding boxes on a random subset of images
num_images_to_plot = 4 # Number of images to randomly select and plot
random_indices = random.sample(range(len(image_files)), num_images_to_plot)


random_image_label_pairs = [(image_files[i], label_files[i]) for i in random_indices]

# random_images = random.sample(image_files, num_images_to_plot)

for image, label in random_image_label_pairs:
    bboxes = parse_bboxes(label)
    draw_bboxes(image, bboxes)
```
